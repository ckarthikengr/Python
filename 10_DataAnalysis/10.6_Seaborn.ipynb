Data Visualization with seaborn


Seaborn is a python visualization library based on matplotlib that provides a high level interface for drawing attractive 
and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code. in this lesson, we will cover the basics of Seaborn, including creating various types of plots and customizing them.

```python
## Basic  plotting with seaborn
import seaborn as sns

```

```python
tips=sns.load_dataset('tips')
tips
```

```python
## create a scatter plot
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter plot of total bill vs tip")
plt.show()
```

```python
## Line plot

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line plot of Total bill by size")
plt.show()
```

```python
## Categorical Plots
## Bar plots

sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar plot of total bill by day')
plt.show()
```

```python
## Box plot
sns.boxplot(x="day",y="total_bill",data=tips)
```

```python
## Viloin plot - check  the distribution
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
## Histograms
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
sns.histplot(tips['total_bill'],bins=10,kde=False)
```

```python
## KDE Plot
sns.kdeplot(tips['total_bill'],shade=True)
```

```python
##Pairplot  - relationship between each feature (It is for numerical variable not categorical)
sns.pairplot(tips)
```

```python
## Heat Map
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
import pandas as pd
sales_df=pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
## Plot total sales by product
plt.figure(figsize=(10,6))
sns.barplot(x='Product Category',y='Total Revenue',data=sales_df,estimator=sum)
plt.title('Total Sales by product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

```python
## Plot total sales by region
plt.figure(figsize=(10,6))
sns.barplot(x='Region',y='Total Revenue',data=sales_df,estimator=sum)
plt.title('Total Sales by region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```
